### CS kernel

```python
from lametlat.utils.funcs import constant_fit
from kernels import hard_kernel_RGR

p_values = [3, 4, 5]
pz_values = [1.83, 2.43, 3.04]
b_ls = [2, 4, 6, 8, 10]

quasi_data = {}
h_data = {}

for i, p in enumerate(p_values):
    quasi_data[p] = np.load(f"../output/dump/renorm_quasi_xdep_p{p}_ratio.jk.npy", allow_pickle=True).item()
    h_data[p] = hard_kernel_RGR(x_ls, pz_gev=pz_values[i], mu=2)

cs_kernel = {f"p{p2}_p{p1}": {} for p1 in p_values for p2 in p_values if p2 > p1}

for b in b_ls:
    for i, p1 in enumerate(p_values):
        for p2 in p_values[i+1:]:
            ratio = np.array(quasi_data[p2][f"b{b}_re"]) / np.array(quasi_data[p1][f"b{b}_re"])
            ratio_gv = jk_ls_avg(ratio)

            cs_kernel[f"p{p2}_p{p1}"][f"b{b}"] = np.log(ratio_gv * h_data[p1] / h_data[p2]) / np.log(p2/p1)

for p1, p2 in [(3, 4), (4, 5), (3, 5)]:
    fig, ax = default_plot()
    for b in b_ls:
        gv_ls = cs_kernel[f"p{p2}_p{p1}"][f"b{b}"]
        ax.fill_between(x_ls, gv.mean(gv_ls)-gv.sdev(gv_ls), gv.mean(gv_ls)+gv.sdev(gv_ls), alpha=0.5, label=f"b={b}")
    plt.title(f"P{p2} / P{p1}")
    plt.legend(loc="upper right", **fs_p)
    plt.xlim([0.3, 0.7])
    plt.ylim([-3, 3])
    plt.tight_layout()
    plt.show()
```
